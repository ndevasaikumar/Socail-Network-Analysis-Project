```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.read_edgelist("dataset/facebook_combined.txt", nodetype=int)
print(G.number_of_nodes(), G.number_of_edges())
G_cc = G.subgraph(max(nx.connected_components(G), key=len)).copy()

```

Base: Simple Graph Plot (for small sampled subgraph)

```python
import random

def draw_sample_subgraph(G, sample_size=200, layout='spring'):
    
    if G.number_of_nodes() <= sample_size:
        H = G.copy()
    else:
        nodes_sample = random.sample(list(G.nodes()), sample_size)
        H = G.subgraph(nodes_sample).copy()

    if layout == 'spring':
        pos = nx.spring_layout(H, seed=42)
    elif layout == 'kamada_kawai':
        pos = nx.kamada_kawai_layout(H)
    elif layout == 'spectral':
        pos = nx.spectral_layout(H)
    else:
        pos = nx.spring_layout(H, seed=42)

    plt.figure(figsize=(10, 10))
    nx.draw_networkx(
        H, pos=pos,
        with_labels=False,
        node_size=20,
        edge_color='lightgray'
    )
    plt.title(f"Sampled Subgraph ({H.number_of_nodes()} nodes, {H.number_of_edges()} edges)")
    plt.axis('off')
    plt.show()

draw_sample_subgraph(G_cc, sample_size=200, layout='spring')
```

Community Detection (Greedy Modularity) + Visualization

```python
from networkx.algorithms.community import greedy_modularity_communities


communities = list(greedy_modularity_communities(G_cc))
print(f"Detected {len(communities)} communities")


node_to_comm = {}
for i, comm in enumerate(communities):
    for node in comm:
        node_to_comm[node] = i


import random

def draw_communities_sample(G, node_to_comm, sample_size=200):
    if G.number_of_nodes() <= sample_size:
        H = G.copy()
    else:
        nodes_sample = random.sample(list(G.nodes()), sample_size)
        H = G.subgraph(nodes_sample).copy()

    pos = nx.spring_layout(H, seed=42)
    comm_indices = [node_to_comm[n] for n in H.nodes()]
    
    plt.figure(figsize=(10, 10))
    nx.draw_networkx_nodes(
        H, pos,
        node_color=comm_indices,
        cmap=plt.cm.tab20,
        node_size=30
    )
    nx.draw_networkx_edges(H, pos, alpha=0.5)
    plt.title("Communities (sampled subgraph)")
    plt.axis('off')
    plt.show()

draw_communities_sample(G_cc, node_to_comm, sample_size=200)
```

PageRank Centrality + Top Nodes + Visualization

```python

pr = nx.pagerank(G_cc)
print("Sample PageRank values for first 5 nodes:")
for n in list(pr.keys())[:5]:
    print(n, pr[n])

top_pr = sorted(pr.items(), key=lambda x: x[1], reverse=True)[:20]
print("\nTop 20 nodes by PageRank:")
for node, score in top_pr:
    print(node, score)


import numpy as np

def draw_pagerank_sample(G, pr, sample_size=200):
    if G.number_of_nodes() <= sample_size:
        H = G.copy()
    else:
        nodes_sample = random.sample(list(G.nodes()), sample_size)
        H = G.subgraph(nodes_sample).copy()

    pos = nx.spring_layout(H, seed=42)

    
    sizes = []
    for n in H.nodes():
        sizes.append(3000 * pr.get(n, 0.0001))

    plt.figure(figsize=(10, 10))
    nx.draw_networkx(
        H, pos,
        node_size=sizes,
        with_labels=False,
        node_color='skyblue',
        edge_color='gray',
        alpha=0.7
    )
    plt.title("Sampled Subgraph with Node Size  PageRank")
    plt.axis('off')
    plt.show()

draw_pagerank_sample(G_cc, pr, sample_size=200)
```

Structural Holes / Brokerage (Constraint & Effective Size)

```python
from networkx.algorithms.structuralholes import constraint, effective_size


constraint_dict = constraint(G_cc)

eff_size_dict = effective_size(G_cc)


nodes_sorted = sorted(
    G_cc.nodes(),
    key=lambda n: (constraint_dict.get(n, 1), -eff_size_dict.get(n, 0))
)

print("Top 10 potential brokers (structural holes):")
for n in nodes_sorted[:10]:
    print(
        f"Node {n}: constraint={constraint_dict.get(n):.4f}, "
        f"effective_size={eff_size_dict.get(n):.4f}"
    )
```

Small-World Properties + Random Graph Comparison

```python

avg_spl = nx.average_shortest_path_length(G_cc)
diameter = nx.diameter(G_cc)

print("Average shortest path length:", avg_spl)
print("Diameter:", diameter)


n = G_cc.number_of_nodes()
m = G_cc.number_of_edges()
p = (2 * m) / (n * (n - 1))  

G_rand = nx.gnp_random_graph(n, p, seed=42)

G_rand_cc = G_rand.subgraph(max(nx.connected_components(G_rand), key=len)).copy()

avg_spl_rand = nx.average_shortest_path_length(G_rand_cc)
diameter_rand = nx.diameter(G_rand_cc)

print("\nRandom graph comparison:")
print("Random graph average shortest path:", avg_spl_rand)
print("Random graph diameter:", diameter_rand)
```

Link Prediction (Jaccard, Adamic-Adar, Preferential Attachment)

```python
from networkx.algorithms.link_prediction import (
    jaccard_coefficient,
    adamic_adar_index,
    preferential_attachment
)


nodes_sample = random.sample(list(G_cc.nodes()), 1000)
H_lp = G_cc.subgraph(nodes_sample).copy()

#
jaccard_preds = list(jaccard_coefficient(H_lp))

jaccard_top = sorted(jaccard_preds, key=lambda x: x[2], reverse=True)[:20]

print("Top 20 potential future links by Jaccard coefficient:")
for u, v, p in jaccard_top:
    print(f"{u} - {v}: {p:.4f}")


aa_preds = list(adamic_adar_index(H_lp))
aa_top = sorted(aa_preds, key=lambda x: x[2], reverse=True)[:20]

print("\nTop 20 potential future links by Adamic-Adar:")
for u, v, p in aa_top:
    print(f"{u} - {v}: {p:.4f}")


pa_preds = list(preferential_attachment(H_lp))
pa_top = sorted(pa_preds, key=lambda x: x[2], reverse=True)[:20]

print("\nTop 20 potential future links by Preferential Attachment:")
for u, v, p in pa_top:
    print(f"{u} - {v}: {p}")
```

Interactive Graph with PyVis

```python
import random
from pyvis.network import Network

def create_interactive_graph(G, output_html="facebook_ego_interactive.html", sample_size=500):
   
    if G.number_of_nodes() > sample_size:
        nodes_sample = random.sample(list(G.nodes()), sample_size)
        H = G.subgraph(nodes_sample).copy()
    else:
        H = G.copy()

    
    net = Network(
        height="750px",
        width="100%",
        bgcolor="#222222",
        font_color="white",
        notebook=False
    )

    net.barnes_hut()

  
    for node, degree in H.degree():
        net.add_node(int(node), label=str(node), value=degree)

   
    for u, v in H.edges():
        net.add_edge(int(u), int(v))

    
    net.save_graph(output_html)

    print(f"\nInteractive graph saved successfully: {output_html}")
    print("Open this file manually in your browser to view it.")


create_interactive_graph(G_cc, "facebook_ego_interactive.html", sample_size=500)
```
